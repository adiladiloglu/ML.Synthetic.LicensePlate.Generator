```python
from PIL import Image, ImageDraw
import os
import json
import Classes.imageAnnotation as ia
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
def getJsonFiles(directory):
    if not os.path.exists(directory):
        return []
    else:
        return [os.path.join(directory,f) for f in os.listdir(directory) if f.endswith(".json")]
jsonFiles = getJsonFiles("/Dataset/Train")
```

```python
def loadAnnotation(jfile):
    with open(jfile) as f:
        return ia.image_annotation_from_dict(json.load(f))
```

```python
def getRandomimageAnnotationPair(json_files):    
    jfile = random.choice(json_files)
    ann = loadAnnotation(jfile)
    directory = os.path.dirname(os.path.realpath(jfile))
    im = Image.open(os.path.join(directory,ann.file))
    return(im,ann)
```

```python
jsonFiles = getJsonFiles("/Dataset/Train")
im,ann = getRandomimageAnnotationPair(jsonFiles)
fig,ax = plt.subplots(1,figsize=(32, 18))
ax.imshow(im)
print("Has {} plates".format(len(ann.annotations)))
for a in ann.annotations:
    if not a.box.is_relative:
        print(a.plate_number)
        print(a.box)
        print(a.segmentation.corners)
        x,y = a.box.left,a.box.top
        w,h = a.box.right-a.box.left,a.box.bottom-a.box.top
        rect = patches.Rectangle((x,y),w,h,linewidth=2,edgecolor='b',facecolor='None')
        ax.add_patch(rect)
        ax.text(a.box.left,a.box.top-10,a.plate_number,fontsize=24, color="white", fontweight="bold",
                bbox={'facecolor': 'black', 'alpha': 0.4, 'pad':5})
        corners = [(int(c.y),int(c.x)) for c in a.segmentation.corners]
        corners.append(corners[0])
        ax.plot(corners[:,0],corners[:,1],color="red", marker="o", )
       
plt.show()
```
